```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
cluster_1 = np.random.randn(50,2)+[2,2]
cluster_2 = np.random.randn(50,2)+[8,8]
cluster_3 = np.random.randn(50,2)+[5,0]
data=np.vstack((cluster_1,cluster_2,cluster_3))
plt.scatter(data[:,0], data[:,1], cmap='rainbow', s=50)
```

```python
from sklearn.cluster import DBSCAN
scaler= StandardScaler()
data_scaled=scaler.fit_transform(data)
dbscan=DBSCAN(eps=0.2, min_samples=3)
clusters= dbscan.fit_predict(data_scaled)
plt.figure(figsize=(8,6))
plt.scatter(data[:,0], data[:,1], c=clusters, cmap='rainbow', s=50)
plt.title("DBSCAN Clustering of 2D Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
print("Cluster labels for each point:")
print(cluster)
```
